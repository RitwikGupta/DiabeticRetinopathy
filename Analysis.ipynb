```python
from pandas import DataFrame
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = DataFrame.from_csv('data', sep=',', index_col=None)
```

```python
data[:3]
```

```python
len(data)
```

```python
temp = [
    [x for x in data['QA']],
    [x for x in data['PRESCREEN']],
    [x for x in data['PT5']],
    [x for x in data['PT8']],
    [x for x in data['PT10']]
]

total = [x for x in zip(*temp)]
training = total[:1036]
testing = total[1036:]
```

```python
actual_data = {
    'data': np.array(training),
    'labels': [x for x in data['CLASS'][:1036]]
}

testing_data = {
    'data': np.array(testing),
    'labels': [x for x in data['CLASS'][1036:]]
}
```

```python
actual_data
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(actual_data['data'], actual_data['labels'])
```

```python
testing[:10], clf.predict(testing)
```

```python
clf.predict([(1,1,3,3,3), (1,1,33,63,93), (1,1,10,20,30)])
```

```python
clf.score(testing_data['data'], testing_data['labels'])
```
